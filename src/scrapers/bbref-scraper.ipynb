```python
import os
import json
import ijson
from pandas.io.json import json_normalize

directory = os.fsencode('results/nba-json/2015-2016')

def clean_team(data, team, container):
    for player in data[team]['players']:
        if (data[team]['players'][player]["MP"] != 0 and 
            'name' in data[team]['players'][player]):
            data[team]['players'][player]['First Name'] = data[team]['players'][player]['name'].split()[0]
            data[team]['players'][player]['Last Name'] = data[team]['players'][player]['name'].split()[1]
            data[team]['players'][player]['Date'] = data['date']
            container.append(data[team]['players'][player])
    

def clean_bbref_logs(directory):
    game_data = []
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        with open(os.path.join(os.fsdecode(directory), filename)) as data_file: 
            if (filename.endswith('.json')):
                print("\n", "Cleaning", filename, "\n")
                data = json.load(data_file)
                clean_team(data, 'home', game_data)
                clean_team(data, 'away', game_data)
            
    df = json_normalize(game_data)
    df.to_csv('foo.csv', index=False)

    
clean_bbref_logs(directory)
        
        
```
